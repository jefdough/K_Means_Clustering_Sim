# Simulating and Testing K-Means Cluster Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
sns.set_style('darkgrid')
```

## Creating the Data

Scikit learn has a function to make "blobs" which are perfect for creating data to test the K Means Clustering algorithm. <br>
http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html


```python
from sklearn.datasets import make_blobs
```

```python
datablobs = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=2.1, random_state=2)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(datablobs[0][:,0], datablobs[0][:,1], c=datablobs[1], cmap='rainbow', edgecolors='grey')
```

## Create Clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(datablobs[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))


ax1.set_title('Original')
ax1.scatter(datablobs[0][:,0],datablobs[0][:,1], c=datablobs[1], cmap='rainbow', edgecolors='grey')

ax2.set_title('K Means')
ax2.scatter(datablobs[0][:,0],datablobs[0][:,1], c=kmeans.labels_, cmap='rainbow', edgecolors='grey')

```

Here we can compare by visualization how well our clusters captured the blobs.
